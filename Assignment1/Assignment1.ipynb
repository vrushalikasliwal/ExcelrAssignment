```python
# Select only numerical columns
numeric_cols = ds.select_dtypes(include=['int64', 'float64']).columns

# Plot histograms
for col in numeric_cols:
    plt.figure()
    plt.hist(ds[col], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
for col in numeric_cols:
    #sns.histplot(ds[col])

    plt.figure()
    sns.histplot(ds['Volume'], bins=30, kde=True, color='skyblue')

    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
## Explain the concept of standardization (z-score normalization).
#Standardization is a technique used to rescale data so that it has
#- A mean (average) of 0
#- A standard deviation of 1
#This transformation helps ensure that all features in a dataset are on the same scale, which is crucial for many statistical and machine learning models
#Why Use Standardization?
#Here’s why it’s so useful:
#- ✅ Equal footing for features: Prevents features with larger scales from dominating models (e.g., income vs. age).
#- ✅ Improves model performance: Especially for algorithms like k-NN, SVM, and PCA that rely on distance calculations.
#- ✅ Outlier detection: Extreme z-scores can highlight anomalies in the data.
#Z Score Formula:- z=x-mu/sigma= (((Age Col Every Value)-Mean)/SD)
```

```python
#Discuss the need for converting categorical data into dummy variables (one-hot encoding).
#Algorithms don’t understand text labels.
#- Assigning arbitrary numbers (e.g., Red = 1, Blue = 2) implies ordinal relationships that don’t exist
#- Avoids misleading order
#- Improves accuracy
#Enables compatibility
#- If a column has many unique categories (e.g., zip codes), one-hot encoding can create too many columns, leading to:
#- Increased memory usage
#- Risk of overfitting
#In such cases, techniques like target encoding or embedding layers might be better.
```

```python
df_encoded = pd.get_dummies(dsCatCols, columns=list(dsCatCols.columns), drop_first=False)
df_encoded 
```
